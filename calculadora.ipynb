Calculadora financiera con Python

```python
user = {'balance':[],
        'income':[],
        'expense':[]}
```

```python
def show_user():
    print(user)
```

```python
# Funcion que permite meter informacion
def add(category,value):
    if category == 'balance':
        user[category].append(value)
    elif category == 'income':
        user[category].append(value)
        user['balance'].append(value + user['balance'][-1])
    elif category == 'expense':
        user[category].append(value)
        user['balance'].append( - value + user['balance'][-1])
```

```python
add('expense',90)
```

```python
user
```

```python
# Funcion que permite ver todos los valores de una categoria
def get(category, value):
    if value == 'all':
        return user[category][:]
    else:
        return user[category][value]
```

```python
get('income',1)
```

```python
# Ver registros
import matplotlib.pyplot as plt

def show(category):
    plt.figure(figsize= (10,6))
    plt.title(f'Timeseries if {category}')
    plt.xlabel('Operation')
    plt.ylabel(f'{category}')
    plt.plot(get(category, 'all'), 'o--')
    plt.show()
    
```

```python
show('balance')
```
